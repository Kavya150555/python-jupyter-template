### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
import matplotlib.pyplot as plt

try:
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]
    colors = ['red', 'green', 'blue', 'purple']

    if len(categories) != len(values) or len(categories) != len(colors):
        raise ValueError("Error: Mismatch in the lengths of 'categories', 'values', and 'colors' lists.")

    plt.figure(figsize=(7, 5))
    plt.bar(categories, values, color=colors)
    plt.title("Bar Chart with Different Colors")
    plt.xlabel("Categories")
    plt.ylabel("Values")
    plt.show()

except ImportError:
    print("Error: Required libraries (matplotlib) are not installed.")
except Exception as e:
    print(f"An error occurred: {e}")
```
