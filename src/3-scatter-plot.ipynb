### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import matplotlib.pyplot as plt

try:
    x = [1, 2, 3, 4, 5]
    y = [2, 4, 6, 8, 10]

    if len(x) != len(y):
        raise ValueError("Error: Length of 'x' and 'y' lists must be the same.")

    plt.figure(figsize=(7, 5))
    plt.scatter(x, y, color="blue", marker="o")

    for i in range(len(x)):
        plt.annotate(f"({x[i]}, {y[i]})", (x[i], y[i]), textcoords="offset points", xytext=(0, 5), ha='center')

    plt.title("Scatter Plot with Annotations")
    plt.xlabel("X values")
    plt.ylabel("Y values")
    plt.grid(True)
    plt.show()

except ImportError:
    print("Error: Required libraries (matplotlib) are not installed.")
except Exception as e:
    print(f"An error occurred: {e}")
```
