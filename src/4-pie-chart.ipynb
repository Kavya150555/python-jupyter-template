### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
```

```python
import matplotlib.pyplot as plt

try:
    labels = ['Python', 'Java', 'C++', 'JavaScript']
    sizes = [40, 25, 20, 15]
    colors = ['gold', 'skyblue', 'lightcoral', 'lightgreen']

    if len(labels) != len(sizes) or len(labels) != len(colors):
        raise ValueError("Error: Length of 'labels', 'sizes', and 'colors' lists must be the same.")

    plt.figure(figsize=(7, 7))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
    plt.title("Pie Chart with Percentages")
    plt.show()

except ImportError:
    print("Error: Required libraries (matplotlib) are not installed.")
except Exception as e:
    print(f"An error occurred: {e}")
```
