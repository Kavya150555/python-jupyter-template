### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
```

```python
import matplotlib.pyplot as plt

def plot_pie_chart():
    try:
        # Data
        labels = ['Python', 'Java', 'C++', 'JavaScript']
        sizes = [40, 25, 20, 15]
        colors = ['blue', 'red', 'green', 'purple']

        # Create pie chart
        plt.figure(figsize=(7, 7))
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, shadow=True, explode=(0.1, 0, 0, 0))

        # Customize plot
        plt.title("Programming Language Usage", fontsize=14)

        # Show plot
        plt.show()

    except Exception as e:
        print(f"An error occurred: {e}")

# Call the function
plot_pie_chart()
```
