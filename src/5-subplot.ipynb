### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import numpy as np
import matplotlib.pyplot as plt

try:
    x = np.linspace(0, 10, 100)
    y = x**2
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]
    scatter_x = [1, 2, 3, 4, 5]
    scatter_y = [2, 4, 6, 8, 10]
    labels = ['Python', 'Java', 'C++', 'JavaScript']
    sizes = [40, 25, 20, 15]

    fig, axes = plt.subplots(2, 2, figsize=(12, 10))

    # Line Plot
    axes[0, 0].plot(x, y, color="b")
    axes[0, 0].set_title("Line Plot")
    axes[0, 0].grid(True)

    # Bar Chart
    axes[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'purple'])
    axes[0, 1].set_title("Bar Chart")

    # Scatter Plot
    axes[1, 0].scatter(scatter_x, scatter_y, color="blue")
    axes[1, 0].set_title("Scatter Plot")

    # Pie Chart
    axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
    axes[1, 1].set_title("Pie Chart")

    plt.tight_layout()
    plt.show()

except ImportError:
    print("Error: Required libraries (matplotlib, numpy) are not installed.")
except Exception as e:
    print(f"An error occurred: {e}")
```
