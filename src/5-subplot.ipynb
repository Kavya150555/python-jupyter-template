### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_subplots():
    try:
        # Create figure and subplots
        fig, axs = plt.subplots(2, 2, figsize=(10, 10))

        # Line plot (Top-left)
        x = np.linspace(0, 10, 100)
        y = x ** 2
        axs[0, 0].plot(x, y, color='b', linewidth=2)
        axs[0, 0].set_title("Line Plot: y = x")
        axs[0, 0].set_xlabel("X Values")
        axs[0, 0].set_ylabel("Y Values")
        axs[0, 0].grid(True, linestyle='--', alpha=0.7)

        # Bar chart (Top-right)
        categories = ['A', 'B', 'C', 'D']
        values = [5, 7, 3, 9]
        colors = ['red', 'blue', 'green', 'purple']
        axs[0, 1].bar(categories, values, color=colors)
        axs[0, 1].set_title("Bar Chart")
        axs[0, 1].set_xlabel("Categories")
        axs[0, 1].set_ylabel("Values")
        axs[0, 1].grid(axis='y', linestyle='--', alpha=0.7)

        # Scatter plot (Bottom-left)
        x_scatter = [1, 2, 3, 4, 5]
        y_scatter = [2, 4, 6, 8, 10]
        axs[1, 0].scatter(x_scatter, y_scatter, color='blue', marker='o')
        axs[1, 0].set_title("Scatter Plot")
        axs[1, 0].set_xlabel("X Values")
        axs[1, 0].set_ylabel("Y Values")
        axs[1, 0].grid(True, linestyle='--', alpha=0.7)

        # Pie chart (Bottom-right)
        labels = ['Python', 'Java', 'C++', 'JavaScript']
        sizes = [40, 25, 20, 15]
        colors = ['blue', 'red', 'green', 'purple']
        axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, shadow=True, explode=(0.1, 0, 0, 0))
        axs[1, 1].set_title("Pie Chart")

        # Adjust layout for clarity
        plt.tight_layout()
        
        # Show plot
        plt.show()

    except Exception as e:
        print(f"An error occurred: {e}")

# Call the function
plot_subplots()
```
